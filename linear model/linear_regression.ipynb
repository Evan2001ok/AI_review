```python
#program of linear regresson
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as sk
#read data
data = pd.read_csv("USA_Housing.csv")
data.head()
```

```python
#list all elements relation
%matplotlib inline
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(10,10))
fig1 = fig.add_subplot(2,2,1)
plt.scatter(data.loc[:,'Avg. Area Income'],data.loc[:,'Price'])
plt.title('Avg. Area Income vs Price')

fig2 = fig.add_subplot(2,2,2)
plt.scatter(data.loc[:,'Avg. Area House Age'],data.loc[:,'Price'])
plt.title('Avg. Area House Age vs Price')

fig3 = fig.add_subplot(2,2,3)
plt.scatter(data.loc[:,'Avg. Area Number of Rooms'],data.loc[:,'Price'])
plt.title('Avg. Area Number of Rooms vs Price')

fig4 = fig.add_subplot(2,2,4)
plt.scatter(data.loc[:,'Avg. Area Number of Bedrooms'],data.loc[:,'Price'])
plt.title('Avg. Area Number of Bedrooms vs Price')

plt.show()
```

```python
#list 2 elements relation and reshape
x = data.loc[:,"Avg. Area Income"]
y = data.loc[:,"Price"]
y.head()
x = np.array(x)
y = np.array(y)
x = x.reshape(-1,1)
y = y.reshape(-1,1)
print(x.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr1 = LinearRegression()
lr1.fit(x,y)
print(lr1.score(x,y))
y1 = lr1.predict(x)
print(y1)
```

```python
#compare MSE and r2_score
from sklearn.metrics import mean_squared_error,r2_score
a = mean_squared_error(y,y1)
b = r2_score(y,y1)
print(a,b)
```

```python
#multi
x_multi = data.drop(['Price','Address'],axis=1)
lr_multi = LinearRegression()
#train model
lr_multi.fit(x_multi,y)
y_multi_predict = lr_multi.predict(x_multi)
from sklearn.metrics import mean_squared_error,r2_score
a = mean_squared_error(y,y_multi_predict)
b = r2_score(y,y_multi_predict)
print(a,b)
fig7 = plt.figure(figsize=(8,5))
plt.scatter(y,y_multi_predict)
plt.show()
fig8 = plt.figure(figsize=(8,5))
plt.scatter(y,y1)
plt.show()
```
