```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('data.csv')
data.head()
```

```python
x = data.drop(['labels'],axis=1)
y = data['labels']
print(x,y)
```

```python
pd.value_counts(y)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
fig1 = plt.figure()
plt.scatter(x['V1'],x['V2'])
plt.show()
```

```python
fig2 = plt.figure()
label1 = plt.scatter(x['V1'][y==0],x['V2'][y==0])
label2 = plt.scatter(x['V1'][y==1],x['V2'][y==1])
label3 = plt.scatter(x['V1'][y==2],x['V2'][y==2])
plt.legend([label1,label2,label3],['0','1','2'])
plt.show()
```

```python
#set model and draw graph
import sklearn
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3,random_state=0)
model.fit(x)

center = model.cluster_centers_
fig3 = plt.figure()
label1 = plt.scatter(x['V1'][y==0],x['V2'][y==0])
label2 = plt.scatter(x['V1'][y==1],x['V2'][y==1])
label3 = plt.scatter(x['V1'][y==2],x['V2'][y==2])
plt.legend([label1,label2,label3],['0','1','2'])
plt.scatter(center[:,0],center[:,1])
plt.show()
```

```python
y_predict_test = model.predict([[80,60]])
print(y_predict_test)
```

```python
y_predict = model.predict(x)
import sklearn.metrics as sm
accuracy = sm.accuracy_score(y,y_predict)
print(accuracy)
```

```python
#predict graph
center = model.cluster_centers_
fig4 = plt.figure()
label1 = plt.scatter(x['V1'][y_predict==0],x['V2'][y_predict==0])
label2 = plt.scatter(x['V1'][y_predict==1],x['V2'][y_predict==1])
label3 = plt.scatter(x['V1'][y_predict==2],x['V2'][y_predict==2])
plt.legend([label1,label2,label3],['0','1','2'])
plt.scatter(center[:,0],center[:,1])
plt.show()
```

```python
#predict graph
fig5 = plt.subplot(121)
label1 = plt.scatter(x['V1'][y==0],x['V2'][y==0])
label2 = plt.scatter(x['V1'][y==1],x['V2'][y==1])
label3 = plt.scatter(x['V1'][y==2],x['V2'][y==2])
plt.legend([label1,label2,label3],['0','1','2'])
plt.scatter(center[:,0],center[:,1])
plt.show()

fig6 = plt.subplot(122)
label1 = plt.scatter(x['V1'][y_predict==0],x['V2'][y_predict==0])
label2 = plt.scatter(x['V1'][y_predict==1],x['V2'][y_predict==1])
label3 = plt.scatter(x['V1'][y_predict==2],x['V2'][y_predict==2])
plt.legend([label1,label2,label3],['0','1','2'])
plt.scatter(center[:,0],center[:,1])
plt.show()
```

```python

y_correct = np.array(y)
print(pd.value_counts(y_correct))
y_correct = []
for i in y_predict:
    if i == 0:
        y_correct.append(2)
    elif i == 1:
        y_correct.append(0)
    else:
        y_correct.append(1)
print(y_correct)
```

```python
#establish a knn model
import sklearn
from sklearn.neighbors import KNeighborsClassifier
import sklearn.metrics as sm
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)

y_predict_knn = knn.predict(x)
accuracy_knn = accuracy_score(y,y_predict_knn)
print(accuracy_knn)
```
