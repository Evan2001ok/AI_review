```python
#program of use mouse draw
import cv2
import numpy as np

drawing = False
mode = 'line'
start_x, start_y = -1, -1
last_x, last_y = -1, -1
color = (0, 255, 0)
thickness = 2
fill = False

img = np.ones((600, 800, 3), np.uint8) * 255
temp_img = img.copy()

# 鼠标回调函数
def draw_shape(event, x, y, flags, param):
    global start_x, start_y, drawing, img, temp_img, last_x, last_y

    if event == cv2.EVENT_LBUTTONDOWN:  # 按下左键开始绘图
        drawing = True
        start_x, start_y = x, y
        last_x, last_y = x, y

        # 如果是自由绘制模式，记录起始点
        if mode == 'free':
            cv2.circle(img, (x, y), thickness//2, color, -1)

    elif event == cv2.EVENT_MOUSEMOVE:  # 鼠标移动
        if drawing:
            if mode == 'free':  # 自由绘制模式
                # 绘制圆点和连接线，获得更平滑的效果
                cv2.circle(img, (x, y), thickness//2, color, -1)
                cv2.line(img, (last_x, last_y), (x, y), color, thickness)
                last_x, last_y = x, y
            else:  # 其他模式：实时预览
                temp_img = img.copy()  # 复制原图，用于预览
                if mode == 'line':
                    cv2.line(temp_img, (start_x, start_y), (x, y), color, thickness)
                elif mode == 'rectangle':
                    if fill:
                        cv2.rectangle(temp_img, (start_x, start_y), (x, y), color, -1)
                    else:
                        cv2.rectangle(temp_img, (start_x, start_y), (x, y), color, thickness)
                elif mode == 'circle':
                    radius = int(np.sqrt((x - start_x)**2 + (y - start_y)**2))
                    if fill:
                        cv2.circle(temp_img, (start_x, start_y), radius, color, -1)
                    else:
                        cv2.circle(temp_img, (start_x, start_y), radius, color, thickness)

    elif event == cv2.EVENT_LBUTTONUP:  # 松开左键完成绘制
        drawing = False
        if mode != 'free':  # 自由绘制模式已经在移动时绘制到原图上了
            if mode == 'line':
                cv2.line(img, (start_x, start_y), (x, y), color, thickness)
            elif mode == 'rectangle':
                if fill:
                    cv2.rectangle(img, (start_x, start_y), (x, y), color, -1)
                else:
                    cv2.rectangle(img, (start_x, start_y), (x, y), color, thickness)
            elif mode == 'circle':
                radius = int(np.sqrt((x - start_x)**2 + (y - start_y)**2))
                if fill:
                    cv2.circle(img, (start_x, start_y), radius, color, -1)
                else:
                    cv2.circle(img, (start_x, start_y), radius, color, thickness)
            # 更新临时图像为最终结果
            temp_img = img.copy()

# 创建窗口和鼠标回调
cv2.namedWindow('Drawing Board')
cv2.setMouseCallback('Drawing Board', draw_shape)

# 显示操作说明
print("绘图工具 - 操作说明:")
print("- 按 'l' 键切换为直线模式")
print("- 按 'r' 键切换为矩形模式")
print("- 按 'c' 键切换为圆形模式")
print("- 按 'f' 键切换为自由绘制模式")
print("- 按 'F' 键切换填充模式（对矩形和圆形有效）")
print("\n颜色控制:")
print("- 按 '1' 键设置为红色")
print("- 按 '2' 键设置为绿色")
print("- 按 '3' 键设置为蓝色")
print("- 按 '4' 键设置为黑色")
print("- 按 '5' 键设置为白色")
print("\n线条控制:")
print("- 按 '+' 键增加线条粗细")
print("- 按 '-' 键减小线条粗细")
print("\n其他操作:")
print("- 按 'C' 键清除画布")
print("- 按 's' 键保存图像")
print("- 按 'q' 键退出程序")

# 主循环
while True:
    # 显示当前图像
    if drawing and mode != 'free':
        cv2.imshow('Drawing Board', temp_img)  # 显示临时图像(预览)
    else:
        cv2.imshow('Drawing Board', img)  # 显示最终图像

    key = cv2.waitKey(1) & 0xFF

    # 切换绘图模式
    if key == ord('l'):
        mode = 'line'
        print("切换为直线模式")
    elif key == ord('r'):
        mode = 'rectangle'
        print("切换为矩形模式")
    elif key == ord('c'):
        mode = 'circle'
        print("切换为圆形模式")
    elif key == ord('f'):
        mode = 'free'
        print("切换为自由绘制模式")
    elif key == ord('F'):
        fill = not fill
        print(f"填充模式: {'开启' if fill else '关闭'}")

    # 颜色设置
    elif key == ord('1'):
        color = (0, 0, 255)  # 红色
        print("切换为红色")
    elif key == ord('2'):
        color = (0, 255, 0)  # 绿色
        print("切换为绿色")
    elif key == ord('3'):
        color = (255, 0, 0)  # 蓝色
        print("切换为蓝色")
    elif key == ord('4'):
        color = (0, 0, 0)  # 黑色
        print("切换为黑色")
    elif key == ord('5'):
        color = (255, 255, 255)  # 白色
        print("切换为白色")

    # 线条粗细
    elif key == ord('+'):
        thickness = min(thickness + 1, 10)
        print(f"线条粗细: {thickness}")
    elif key == ord('-'):
        thickness = max(thickness - 1, 1)
        print(f"线条粗细: {thickness}")

    # 清除画布
    elif key == ord('C'):
        img = np.ones((600, 800, 3), np.uint8) * 255
        temp_img = img.copy()
        print("画布已清除")

    # 保存图像
    elif key == ord('s'):
        cv2.imwrite('drawing.png', img)
        print("图像已保存为 drawing.png")

    # 退出程序
    elif key == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 加载图片
img = cv2.imread('1.jpg')
height, width = img.shape[:2]

# 1. 平移矩阵 - 向右和向下移动50像素
translation_matrix = np.float32([
    [1, 0, 50],  # x方向平移50
    [0, 1, 50]   # y方向平移50
])
translated = cv2.warpAffine(img, translation_matrix, (width, height))

# 2. 旋转矩阵 - 围绕图像中心旋转45度
center = (width // 2, height // 2)
angle = 45
scale = 1.0
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(img, rotation_matrix, (width, height))

# 3. 缩放矩阵 - 水平方向缩放为原来的1.5倍，垂直方向为原来的0.8倍
scaling_matrix = np.float32([
    [1.5, 0, 0],
    [0, 0.8, 0]
])
scaled = cv2.warpAffine(img, scaling_matrix, (width, height))

# 4. 仿射变换 - 将图像的三个点映射到新的三个点
src_points = np.float32([[0, 0], [width-1, 0], [0, height-1]])
dst_points = np.float32([[0, 0], [width-1, 0], [width//4, height-1]])
affine_matrix = cv2.getAffineTransform(src_points, dst_points)
affine = cv2.warpAffine(img, affine_matrix, (width, height))

# 5. 透视变换 - 矫正倾斜的矩形
src_perspective = np.float32([
    [0, 0],
    [width-1, 0],
    [width-1, height-1],
    [0, height-1]
])
dst_perspective = np.float32([
    [width*0.1, height*0.1],
    [width*0.9, height*0.1],
    [width*0.9, height*0.9],
    [width*0.1, height*0.9]
])
perspective_matrix = cv2.getPerspectiveTransform(src_perspective, dst_perspective)
perspective = cv2.warpPerspective(img, perspective_matrix, (width, height))

# 显示结果
cv2.imshow('原图', img)
cv2.imshow('平移', translated)
cv2.imshow('旋转', rotated)
cv2.imshow('缩放', scaled)
cv2.imshow('仿射变换', affine)
cv2.imshow('透视变换', perspective)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
