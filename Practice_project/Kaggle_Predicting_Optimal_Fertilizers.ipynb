```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pickle

# 加载数据
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
orig_df = pd.read_csv('Fertilizer_Prediction.csv')

print("数据基本信息:")
print(f"训练集形状: {train_df.shape}")
print(f"测试集形状: {test_df.shape}")
print(f"原始数据集形状: {orig_df.shape}")

# 1. 分离特征和目标变量
X_train_full = train_df.drop(['id', 'Fertilizer Name'], axis=1)
y_train_full = train_df['Fertilizer Name']
X_test = test_df.drop(['id'], axis=1)

# 原始数据处理
X_orig_full = orig_df.drop(['Fertilizer Name'], axis=1)
y_orig_full = orig_df['Fertilizer Name']

# 2. 处理分类特征 - 使用Label Encoding
label_encoders = {}
categorical_columns = ['Soil Type', 'Crop Type']

for column in categorical_columns:
    le = LabelEncoder()

    # 合并所有数据的该列，确保编码器见过所有可能的类别
    all_categories = pd.concat([
        X_train_full[column],
        X_test[column],
        X_orig_full[column]
    ]).dropna()

    # 拟合编码器
    le.fit(all_categories)

    # 分别转换各个数据集
    X_train_full[column] = le.transform(X_train_full[column])
    X_test[column] = le.transform(X_test[column])
    X_orig_full[column] = le.transform(X_orig_full[column])

    # 保存编码器
    label_encoders[column] = le

    print(f"{column} 类别数: {len(le.classes_)}")
    print(f"{column} 类别: {le.classes_}")

# 3. 对目标变量进行编码
le_target = LabelEncoder()

# 合并训练集和原始数据的目标变量进行拟合
all_targets = pd.concat([y_train_full, y_orig_full]).dropna()
le_target.fit(all_targets)

# 转换目标变量
y_train_full_encoded = le_target.transform(y_train_full)
y_orig_full_encoded = le_target.transform(y_orig_full)

print(f"\n目标变量类别数: {len(le_target.classes_)}")
print(f"目标变量类别: {le_target.classes_}")

# 4. 先分割训练集和验证集（只使用比赛提供的训练数据）
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_train_full_encoded
)

# 5. 可选：将原始数据多次复制后添加到训练集（不影响验证集）
print("\n是否使用原始数据增强训练集？")
use_original_data = True  # 可以设置为 False 如果不想使用
orig_copies = 5  # 原始数据复制份数

if use_original_data:
    # 创建原始数据的多个副本
    X_orig_copies = []
    y_orig_copies = []

    for i in range(orig_copies):
        X_orig_copies.append(X_orig_full.copy())
        y_orig_copies.append(y_orig_full_encoded.copy())
        print(f"创建原始数据副本 {i+1}/{orig_copies}")

    # 合并所有原始数据副本
    X_orig_multiplied = pd.concat(X_orig_copies, axis=0, ignore_index=True)
    y_orig_multiplied = np.concatenate(y_orig_copies)

    # 将增强后的原始数据添加到训练集
    X_train_enhanced = pd.concat([X_train, X_orig_multiplied], axis=0, ignore_index=True)
    y_train_enhanced = np.concatenate([y_train, y_orig_multiplied])

    # 更新训练集
    X_train = X_train_enhanced
    y_train = y_train_enhanced

    print(f"已将原始数据复制 {orig_copies} 份并添加到训练集")
    print(f"原始数据总样本数: {len(X_orig_multiplied)} (原始: {len(X_orig_full)} × {orig_copies})")
    print("验证集保持不变")

# 打印最终数据形状
print(f"\n最终数据形状:")
print(f"训练集: {X_train.shape}")
print(f"验证集: {X_val.shape}")
print(f"测试集: {X_test.shape}")

# 检查数据类型
print(f"\n特征列: {X_train.columns.tolist()}")
print(f"数值特征统计:")
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
print(X_train[numeric_cols].describe())

# 6. 保存预处理对象
preprocessing_objects = {
    'label_encoders': label_encoders,
    'target_encoder': le_target,
    'feature_columns': X_train.columns.tolist(),
    'target_classes': le_target.classes_
}

# 保存到文件
with open('preprocessing_objects.pkl', 'wb') as f:
    pickle.dump(preprocessing_objects, f)

print("\n预处理对象已保存到 'preprocessing_objects.pkl'")

# 7. 检查是否有缺失值
print(f"\n缺失值检查:")
print(f"训练集缺失值: {X_train.isnull().sum().sum()}")
print(f"验证集缺失值: {X_val.isnull().sum().sum()}")
print(f"测试集缺失值: {X_test.isnull().sum().sum()}")

# 8. 显示类别分布
import numpy as np
print(f"\n验证集目标变量分布:")
unique_val, counts_val = np.unique(y_val, return_counts=True)
for cls_idx, count in zip(unique_val, counts_val):
    cls_name = le_target.classes_[cls_idx]
    print(f"{cls_name}: {count} ({count/len(y_val)*100:.1f}%)")

print(f"\n训练集目标变量分布 (包含原始数据增强):")
unique_train, counts_train = np.unique(y_train, return_counts=True)
for cls_idx, count in zip(unique_train, counts_train):
    cls_name = le_target.classes_[cls_idx]
    print(f"{cls_name}: {count} ({count/len(y_train)*100:.1f}%)")
```

```python
def xgb_gpu_evaluation(X_train, X_val, y_train, y_val, cv_folds=5):
    from xgboost import XGBClassifier
    import numpy as np
    from sklearn.model_selection import cross_val_score, StratifiedKFold

    # MAP@3计算函数
    def mapk(actual, predicted, k=3):
        def apk(a, p, k):
            if a in p[:k]:
                return 1.0 / (p[:k].index(a) + 1)
            return 0.0
        return np.mean([apk(a, p, k) for a, p in zip(actual, predicted)])

    # 创建模型
    # 把模型参数改为之前最佳的
    xgb = XGBClassifier(
      n_estimators=400,
      learning_rate=0.05,
      max_depth=7,
      subsample=0.85,
      colsample_bytree=0.85,
      reg_alpha=0.05,
      reg_lambda=0.5,
      random_state=42,
      tree_method='hist',
      device='cuda',
      enable_categorical=True)

    # 交叉验证
    cv = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=42)
    cv_scores = cross_val_score(xgb, X_train, y_train, cv=cv, scoring='accuracy')

    # 训练最终模型
    xgb.fit(X_train, y_train)

    # 验证集评估
    val_accuracy = xgb.score(X_val, y_val)

    # 计算MAP@3
    y_proba = xgb.predict_proba(X_val)
    top3_preds = np.argsort(y_proba, axis=1)[:, -3:][:, ::-1]
    map3_score = mapk(y_val.tolist(), top3_preds.tolist(), k=3)

    # 输出结果
    print(f"=== XGBoost结果 ===")
    print(f"交叉验证准确率: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
    print(f"验证集准确率: {val_accuracy:.4f}")
    print(f"验证集MAP@3: {map3_score:.4f}")

    return xgb, val_accuracy, map3_score

# 使用
model, accuracy, map3 = xgb_gpu_evaluation(X_train, X_val, y_train, y_val)
```

```python
#单个最佳模型
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

# 加载数据
print("加载数据...")
train = pd.read_csv('train.csv').drop(columns=['id'])
test = pd.read_csv('test.csv').drop(columns=['id'])
original = pd.read_csv('Fertilizer_Prediction.csv')

# 数据增强：复制5次
original = pd.concat([original] * 6, ignore_index=True)

print(f"训练集: {train.shape}")
print(f"测试集: {test.shape}")
print(f"原始数据: {original.shape}")

# 特征工程
def create_features(df):
    """创建分箱和零值特征"""
    df = df.copy()

    # 数值列分箱
    num_cols = ['Nitrogen', 'Phosphorous', 'Potassium', 'Temparature', 'Humidity', 'Moisture']
    for col in [c for c in num_cols if c in df.columns]:
        df[f"{col}_Binned"] = df[col].astype("category")

    # 零值特征
    for col in ['Phosphorous', 'Potassium']:
        if col in df.columns:
            df[f"{col}_BinnedZero"] = df[col].apply(lambda x: 'ZERO' if x == 0 else str(x)).astype("category")

    return df

train['const'] = 0
test['const'] = 0
original['const'] = 0

# 应用特征工程
print("\n应用特征工程...")
train = create_features(train)
test = create_features(test)
original = create_features(original)

# 编码类别特征
for col in ['Soil Type', 'Crop Type']:
    le = LabelEncoder()
    all_values = pd.concat([train[col], test[col], original[col]]).unique()
    le.fit(all_values)

    train[col] = pd.Categorical(le.transform(train[col]))
    test[col] = pd.Categorical(le.transform(test[col]))
    original[col] = pd.Categorical(le.transform(original[col]))

# 创建交互特征
print("创建交互特征...")
for df in [train, test, original]:
    df['Soil_Crop'] = (df['Soil Type'].astype(str) + "_" + df['Crop Type'].astype(str)).astype("category")

# 准备训练数据
print("\n准备训练数据...")
target_encoder = LabelEncoder()
y_train = target_encoder.fit_transform(train['Fertilizer Name'])
y_original = target_encoder.transform(original['Fertilizer Name'])

feature_cols = [col for col in test.columns if col != 'Fertilizer Name']
print(f"特征数量: {len(feature_cols)}")

# 分割数据集 - 使用种子123（最佳表现）
X_tr, X_val, y_tr, y_val = train_test_split(
    train[feature_cols], y_train,
    test_size=0.2, stratify=y_train,
    random_state=123  # 修改为123
)

# 合并原始数据
X_tr = pd.concat([X_tr, original[feature_cols]], ignore_index=True)
y_tr = np.concatenate([y_tr, y_original])

print(f"训练集: {X_tr.shape}, 验证集: {X_val.shape}")

# XGBoost训练
params = {
    'objective': 'multi:softprob',
    'num_class': len(target_encoder.classes_),
    'seed': 123,  # 修改为123
    'tree_method': 'hist',
    'device': 'cuda',
    'eval_metric': 'mlogloss',
    'max_depth': 8,
    'learning_rate': 0.04,
    'min_child_weight': 2,
    'alpha': 5.5,
    'reg_lambda': 3.0,
    'subsample': 0.8,
    'colsample_bytree': 0.3,
    'max_bins': 128,
    'gamma': 0
}

# 创建DMatrix
dtrain = xgb.DMatrix(X_tr, label=y_tr, enable_categorical=True)
dvalid = xgb.DMatrix(X_val, label=y_val, enable_categorical=True)
dtest = xgb.DMatrix(test[feature_cols], enable_categorical=True)

# 训练模型
print("\n开始训练...")
print("使用种子123 - 在多种子测试中表现最佳 (MAP@3=0.377738)")
model = xgb.train(
    params, dtrain,
    num_boost_round=10000,
    evals=[(dtrain, 'train'), (dvalid, 'valid')],
    early_stopping_rounds=50,
    verbose_eval=100
)

print(f"\n最佳迭代: {model.best_iteration}")
print(f"最佳验证分数: {model.best_score:.6f}")

# MAP@3评估
def mapk(actual, predicted, k=3):
    """计算MAP@K分数"""
    scores = []
    for a, p in zip(actual, predicted):
        score, hits = 0.0, 0
        for i, pred in enumerate(p[:k]):
            if pred in a:
                hits += 1
                score += hits / (i + 1.0)
        scores.append(score / min(len(a), k))
    return np.mean(scores)

# 验证集评估
val_pred = model.predict(dvalid)
top3_preds = np.argsort(val_pred, axis=1)[:, -3:][:, ::-1]
map3_score = mapk([[label] for label in y_val], top3_preds)
print(f"\n验证集 MAP@3: {map3_score:.6f}")

# 特征重要性
print("\n=== Top 20 重要特征 ===")
importance = model.get_score(importance_type='gain')
for i, (feat, score) in enumerate(sorted(importance.items(), key=lambda x: x[1], reverse=True)[:20], 1):
    print(f"{i:2d}. {feat:25s}: {score:8.2f}")

# 生成预测
print("\n生成预测...")
test_pred = model.predict(dtest)
top3_test = np.argsort(test_pred, axis=1)[:, -3:][:, ::-1]
top3_labels = target_encoder.inverse_transform(top3_test.ravel()).reshape(top3_test.shape)

# 保存结果
submission = pd.DataFrame({
    'id': pd.read_csv('test.csv')['id'],
    'Fertilizer Name': [' '.join(row) for row in top3_labels]
})
submission.to_csv('best_seed_123_submission.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

# 加载数据
print("加载数据...")
train = pd.read_csv('train.csv').drop(columns=['id'])
test = pd.read_csv('test.csv').drop(columns=['id'])
original = pd.read_csv('Fertilizer_Prediction.csv')

# 数据增强：复制6次（基于你的最新设置）
original = pd.concat([original] * 6, ignore_index=True)

print(f"训练集: {train.shape}")
print(f"测试集: {test.shape}")
print(f"原始数据: {original.shape}")

# 特征工程
def create_features(df):
    """创建分箱和零值特征"""
    df = df.copy()

    # 数值列分箱
    num_cols = ['Nitrogen', 'Phosphorous', 'Potassium', 'Temparature', 'Humidity', 'Moisture']
    for col in [c for c in num_cols if c in df.columns]:
        df[f"{col}_Binned"] = df[col].astype("category")

    # 零值特征
    for col in ['Phosphorous', 'Potassium']:
        if col in df.columns:
            df[f"{col}_BinnedZero"] = df[col].apply(lambda x: 'ZERO' if x == 0 else str(x)).astype("category")

    return df

# 添加常数特征
train['const'] = 0
test['const'] = 0
original['const'] = 0

# 应用特征工程
print("\n应用特征工程...")
train = create_features(train)
test = create_features(test)
original = create_features(original)

# 编码类别特征
for col in ['Soil Type', 'Crop Type']:
    le = LabelEncoder()
    all_values = pd.concat([train[col], test[col], original[col]]).unique()
    le.fit(all_values)

    train[col] = pd.Categorical(le.transform(train[col]))
    test[col] = pd.Categorical(le.transform(test[col]))
    original[col] = pd.Categorical(le.transform(original[col]))

# 创建交互特征
print("创建交互特征...")
for df in [train, test, original]:
    df['Soil_Crop'] = (df['Soil Type'].astype(str) + "_" + df['Crop Type'].astype(str)).astype("category")

# 准备训练数据
print("\n准备训练数据...")
target_encoder = LabelEncoder()
y_train = target_encoder.fit_transform(train['Fertilizer Name'])
y_original = target_encoder.transform(original['Fertilizer Name'])

feature_cols = [col for col in test.columns if col != 'Fertilizer Name']
print(f"特征数量: {len(feature_cols)}")

# XGBoost参数
params = {
    'objective': 'multi:softprob',
    'num_class': len(target_encoder.classes_),
    'seed': 123,
    'tree_method': 'hist',
    'device': 'cuda',
    'eval_metric': 'mlogloss',
    'max_depth': 8,
    'learning_rate': 0.04,
    'min_child_weight': 2,
    'alpha': 5.5,
    'reg_lambda': 3.0,
    'subsample': 0.8,
    'colsample_bytree': 0.3,
    'max_bins': 128,
    'gamma': 0
}

# MAP@3评估函数
def mapk(actual, predicted, k=3):
    """计算MAP@K分数"""
    scores = []
    for a, p in zip(actual, predicted):
        score, hits = 0.0, 0
        for i, pred in enumerate(p[:k]):
            if pred in a:
                hits += 1
                score += hits / (i + 1.0)
        scores.append(score / min(len(a), k))
    return np.mean(scores)

# 10折交叉验证
n_folds = 10
skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=123)

# 存储每个fold的预测
oof_predictions = np.zeros((len(train), len(target_encoder.classes_)))
test_predictions = np.zeros((len(test), len(target_encoder.classes_)))
fold_scores = []
best_iterations = []

print(f"\n开始 {n_folds} 折交叉验证训练...")
print("="*60)

# 对每个fold训练模型
for fold_num, (train_idx, val_idx) in enumerate(skf.split(train[feature_cols], y_train)):
    print(f"\nFold {fold_num + 1}/{n_folds}")
    print("-"*40)

    # 分割数据
    X_fold_train = train[feature_cols].iloc[train_idx]
    X_fold_val = train[feature_cols].iloc[val_idx]
    y_fold_train = y_train[train_idx]
    y_fold_val = y_train[val_idx]

    # 添加原始数据到训练集
    X_fold_train_with_original = pd.concat([X_fold_train, original[feature_cols]], ignore_index=True)
    y_fold_train_with_original = np.concatenate([y_fold_train, y_original])

    print(f"Fold训练集: {X_fold_train_with_original.shape}")
    print(f"Fold验证集: {X_fold_val.shape}")

    # 创建DMatrix
    dtrain = xgb.DMatrix(X_fold_train_with_original, label=y_fold_train_with_original, enable_categorical=True)
    dvalid = xgb.DMatrix(X_fold_val, label=y_fold_val, enable_categorical=True)
    dtest = xgb.DMatrix(test[feature_cols], enable_categorical=True)

    # 训练模型
    model = xgb.train(
        params,
        dtrain,
        num_boost_round=10000,
        evals=[(dtrain, 'train'), (dvalid, 'valid')],
        early_stopping_rounds=50,
        verbose_eval=500  # 减少输出
    )

    print(f"最佳迭代: {model.best_iteration}")
    best_iterations.append(model.best_iteration)

    # 保存验证集预测（Out-of-Fold）
    oof_predictions[val_idx] = model.predict(dvalid)

    # 累加测试集预测
    test_predictions += model.predict(dtest) / n_folds

    # 计算fold的MAP@3
    fold_top3 = np.argsort(oof_predictions[val_idx], axis=1)[:, -3:][:, ::-1]
    fold_map3 = mapk([[y] for y in y_fold_val], fold_top3)
    fold_scores.append(fold_map3)

    print(f"Fold {fold_num + 1} 验证集 MAP@3: {fold_map3:.6f}")

    # 清理内存
    del model, dtrain, dvalid

# 计算整体结果
print("\n" + "="*60)
print("交叉验证结果汇总:")
print("="*60)

# 各fold分数
print("\n各Fold MAP@3分数:")
for i, score in enumerate(fold_scores):
    print(f"Fold {i+1:2d}: {score:.6f}")

print(f"\n统计信息:")
print(f"平均 MAP@3: {np.mean(fold_scores):.6f}")
print(f"标准差: {np.std(fold_scores):.6f}")
print(f"最高: {np.max(fold_scores):.6f} (Fold {np.argmax(fold_scores) + 1})")
print(f"最低: {np.min(fold_scores):.6f} (Fold {np.argmin(fold_scores) + 1})")

# 计算完整的OOF分数
oof_top3 = np.argsort(oof_predictions, axis=1)[:, -3:][:, ::-1]
final_oof_score = mapk([[y] for y in y_train], oof_top3)
print(f"\n整体 OOF MAP@3: {final_oof_score:.6f}")

# 最佳迭代统计
print(f"\n最佳迭代统计:")
print(f"平均: {np.mean(best_iterations):.0f}")
print(f"最小: {np.min(best_iterations)}")
print(f"最大: {np.max(best_iterations)}")

# 生成最终预测
print("\n生成测试集预测...")
top3_test = np.argsort(test_predictions, axis=1)[:, -3:][:, ::-1]
top3_labels = target_encoder.inverse_transform(top3_test.ravel()).reshape(top3_test.shape)

# 保存提交文件
submission = pd.DataFrame({
    'id': pd.read_csv('test.csv')['id'],
    'Fertilizer Name': [' '.join(row) for row in top3_labels]
})
submission.to_csv('kfold_10_submission.csv', index=False)

print("\n完成！")
print(f"使用 {n_folds} 折交叉验证集成")
print(f"预期分数基于 OOF: {final_oof_score:.6f}")
print("提交文件已保存: kfold_10_submission.csv")

# 保存OOF预测和测试预测供后续分析
np.save('oof_predictions_10fold.npy', oof_predictions)
np.save('test_predictions_10fold.npy', test_predictions)
print("\n额外保存:")
print("- OOF预测: oof_predictions_10fold.npy")
print("- 测试预测: test_predictions_10fold.npy")
```
